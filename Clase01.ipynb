# PYTHON

```python
from sympy import symbols
from sympy.logic.boolalg import Or, And, Not, Implies, Equivalent

# Definimos las proposiciones
P, Q, R = symbols('P Q R')

# Ejemplos de expresiones lógicas
expr1 = Or(P, Q)            # P OR Q
expr2 = And(P, Not(Q))      # P AND NOT Q
expr3 = Implies(P, Q)       # P IMPLIES Q
expr4 = Equivalent(P, Q)    # P EQUIVALENT Q

# Mostrar las expresiones
print("P OR Q:", expr1)
print("P AND NOT Q:", expr2)
print("P IMPLIES Q:", expr3)
print("P EQUIVALENT Q:", expr4)

# Evaluación de expresiones
valores = {P: True, Q: False}

print("Evaluación de P OR Q con P=True, Q=False:", expr1.subs(valores))
print("Evaluación de P AND NOT Q con P=True, Q=False:", expr2.subs(valores))
print("Evaluación de P IMPLIES Q con P=True, Q=False:", expr3.subs(valores))
print("Evaluación de P EQUIVALENT Q con P=True, Q=False:", expr4.subs(valores))
```

```python
# Definición de conjuntos
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Unión
union = A | B
print("Unión de A y B:", union)

# Intersección
interseccion = A & B
print("Intersección de A y B:", interseccion)

# Diferencia
diferencia = A - B
print("Diferencia de A y B (A - B):", diferencia)

# Diferencia simétrica
diferencia_simetrica = A ^ B
print("Diferencia simétrica de A y B:", diferencia_simetrica)

# Subconjunto
es_subconjunto = A <= B
print("A es subconjunto de B:", es_subconjunto)

# Superconjunto
es_superconjunto = A >= B
print("A es superconjunto de B:", es_superconjunto)
```
