# Utilidad de las Funciones

Las funciones son herramientas fundamentales en matemáticas y otras disciplinas por las siguientes razones:

Modelado de relaciones: Permiten describir relaciones entre conjuntos de datos, como la relación entre la velocidad y el tiempo en física.

Predicción: Las funciones ayudan a predecir valores desconocidos a partir de datos conocidos.

Transformación: Son utilizadas para transformar datos en estadísticas, gráficos y análisis.

Simplificación de problemas: Facilitan la resolución de problemas complejos al descomponerlos en relaciones más simples.

```python
import numpy as np
import matplotlib.pyplot as plt

class FuncionLineal:
    def __init__(self, pendiente, intercepto):
        self.m = pendiente
        self.b = intercepto

    def valor(self, x):
        """Calcula el valor de la función lineal para un valor dado de x."""
        return self.m * x + self.b

    def derivada(self):
        """Retorna la pendiente de la función lineal, que es la derivada."""
        return self.m

    def graficar(self, x_range):
        """Grafica la función lineal en un rango dado de valores de x."""
        x = np.linspace(x_range[0], x_range[1], 400)
        y = self.valor(x)
        plt.plot(x, y, label=f'f(x) = {self.m}x + {self.b}')
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.title('Gráfica de la Función Lineal')
        plt.axhline(0, color='black',linewidth=0.5)
        plt.axvline(0, color='black',linewidth=0.5)
        plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
        plt.legend()
        plt.show()

# Ejemplo de uso
pendiente = 2
intercepto = -3
funcion = FuncionLineal(pendiente, intercepto)

# Calcular el valor de la función para x = 1
x_valor = 1
print(f'f({x_valor}) = {funcion.valor(x_valor)}')

# Calcular la derivada de la función
print(f'Derivada de la función: {funcion.derivada()}')

# Graficar la función en el rango de -10 a 10
funcion.graficar((-10, 10))
```

```python
def is_injective(function, domain):
    """Check if a function is injective (one-to-one)"""
    codomain = set()
    for x in domain:
        y = function(x)
        if y in codomain:
            return False
        codomain.add(y)
    return True

def is_surjective(function, domain, codomain):
    """Check if a function is surjective (onto)"""
    mapped_codomain = set()
    for x in domain:
        y = function(x)
        mapped_codomain.add(y)
    return mapped_codomain == codomain

def is_bijective(function, domain, codomain):
    """Check if a function is bijective (both injective and surjective)"""
    return is_injective(function, domain) and is_surjective(function, domain, codomain)

# Example functions
def function1(x):
    return 2 * x

def function2(x):
    return x ** 2

def function3(x):
    return x + 1

# Example domains and codomains
domain = [1, 2, 3, 4]
codomain1 = {2, 4, 6, 8}
codomain2 = {1, 4, 9, 16}
codomain3 = {2, 3, 4, 5}

# Check properties
print("Function 1 (2*x) is injective:", is_injective(function1, domain))
print("Function 1 (2*x) is surjective:", is_surjective(function1, domain, codomain1))
print("Function 1 (2*x) is bijective:", is_bijective(function1, domain, codomain1))

print("Function 2 (x^2) is injective:", is_injective(function2, domain))
print("Function 2 (x^2) is surjective:", is_surjective(function2, domain, codomain2))
print("Function 2 (x^2) is bijective:", is_bijective(function2, domain, codomain2))

print("Function 3 (x+1) is injective:", is_injective(function3, domain))
print("Function 3 (x+1) is surjective:", is_surjective(function3, domain, codomain3))
print("Function 3 (x+1) is bijective:", is_bijective(function3, domain, codomain3))
```

```python
def f(x):
    return x + 2

def g(x):
    return x * 3

def compose(f, g):
    return lambda x: f(g(x))

# Crear una función compuesta de f y g
h = compose(f, g)

# Probar la función compuesta
print(h(5))  # Esto debería imprimir 17
```
