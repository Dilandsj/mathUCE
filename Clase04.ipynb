# Python

```
import numpy as np

# Definir la función
def funcion(x):
    return 2 * x + 3

# Rango de valores de entrada (dominio)
valores_entrada = np.linspace(-10, 10, 100)

# Calcular la imagen (valores de salida)
valores_salida = funcion(valores_entrada)

# Mostrar resultados
print("Valores de entrada:", valores_entrada)
print("Valores de salida (imagen):", valores_salida)

# Para la preimagen, resolvemos la ecuación inversa
def funcion_inversa(y):
    return (y - 3) / 2

# Valores de salida para los que queremos encontrar la preimagen
valores_salida_deseados = np.linspace(-10, 10, 100)

# Calcular la preimagen
valores_preimagen = funcion_inversa(valores_salida_deseados)

# Mostrar resultados
print("Valores de salida deseados:", valores_salida_deseados)
print("Valores de preimagen:", valores_preimagen)
```
